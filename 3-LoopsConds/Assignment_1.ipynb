### Рисуем черепашкой

```python
import jturtle as t

def poly(x,n):
  for _ in range(n):
    t.forward(x)
    t.right(360/n)

for i in range(36):
  poly(100,40)
  t.right(36)
t.done()
```

```python
import jturtle as t

# Настройка
t.setup(400, 400)
t.title("Простой домик")
t.bgcolor("white")
t.speed(3)
t.hideturtle()

# Основание дома (квадрат)
t.penup()
t.goto(-100, -100)
t.pendown()

t.color("black", "lightblue")
t.begin_fill()
for _ in range(4):
    t.forward(200)
    t.left(90)
t.end_fill()

# Крыша (треугольник)
t.penup()
t.goto(-120, 100)
t.pendown()

t.color("black", "red")
t.begin_fill()
t.goto(0, 200)
t.goto(120, 100)
t.goto(-120, 100)
t.end_fill()

# Дверь
t.penup()
t.goto(-30, -100)
t.pendown()

t.color("black", "brown")
t.begin_fill()
for _ in range(2):
    t.forward(60)
    t.left(90)
    t.forward(100)
    t.left(90)
t.end_fill()

# Ручка двери
t.penup()
t.goto(15, -50)
t.pendown()
t.color("black")
t.dot(8)

# Окно
t.penup()
t.goto(40, 0)
t.pendown()

t.color("black", "white")
t.begin_fill()
for _ in range(4):
    t.forward(50)
    t.left(90)
t.end_fill()

t.penup()
t.goto(40, 25)
t.pendown()
t.forward(50)

t.penup()
t.goto(65, 0)
t.pendown()
t.left(90)
t.forward(50)

t.done()
```

### Рисуем с помощью координатной сетки

```python
for i in range(11):
  t.line((i*10,0),(i*10,100))
for i in range(11):
  t.line((0,i*10),(100,i*10))
t.done()
```
